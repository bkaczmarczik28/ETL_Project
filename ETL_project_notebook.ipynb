```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

# Extract CSV files into a dataframe

```python
csv_airbnb = "Resources/AB_NYC_2019.csv"
airbnb_df = pd.read_csv(csv_airbnb)
airbnb_df.head()
```

```python
csv_properties = "Resources/nyc-rolling-sales.csv"
nycprops_df = pd.read_csv(csv_properties)
nycprops_df.head()
```

# Transform AirBNB DataFrame

```python
airbnb_cols = ["id","neighbourhood_group", "neighbourhood", "latitude", "longitude", "room_type", "price", "number_of_reviews", "calculated_host_listings_count"]
airbnb_transformed= airbnb_df[airbnb_cols].copy()
airbnb_transformed
```

```python
airbnb_transformed = airbnb_transformed.rename(columns={"calculated_host_listings_count": "listings_per_host",
                                                        "id":"airbnb_id",
                                                        "price":"price_$"
                                                       })
airbnb_transformed
```

```python
#drop duplicates
airbnb_transformed.drop_duplicates("airbnb_id", inplace=True)
airbnb_transformed
```

```python
# Set index
airbnb_transformed.set_index("airbnb_id", inplace=True)

airbnb_transformed.head()
```

# Transform NYC Property Dataframe

```python
nycprops_cols=['NEIGHBORHOOD','BUILDING CLASS CATEGORY','BLOCK','LOT', 'ADDRESS', 'ZIP CODE', 'TOTAL UNITS', 'GROSS SQUARE FEET', 'SALE PRICE', 'SALE DATE']
nycprops_transformed= nycprops_df[nycprops_cols].copy()
nycprops_transformed.head()
```

```python
nycprops_transformed = nycprops_transformed.rename(columns={"NEIGHBORHOOD": "neighborhood",
                                                            "BUILDING CLASS CATEGORY" :"building_class_category",
                                                            "BLOCK":"block",
                                                            "LOT" : "lot",
                                                            "ADDRESS":"address",
                                                            "ZIP CODE":"zip_code",
                                                            "TOTAL UNITS":"total_units",
                                                            "GROSS SQUARE FEET":"gross_sq_feet",
                                                            "SALE PRICE":"sale_price_$",
                                                            "SALE DATE":"sale_date"
                                                       })
nycprops_transformed.head()
```

```python
# reformat sales date to just include the date
new = nycprops_transformed["sale_date"].str.split(" ", n = 1, expand = True) 
new.head()
```

```python
nycprops_transformed['sale_date']=new[0]
nycprops_transformed.head()
```

```python
#check size of dataframe
nycprops_transformed.size
```

```python
#drop any listings without a sales price
nycprops_transformed.dropna(subset=['sale_price_$'])
nycprops_transformed.size
```

```python
#replace empty cells with - with 
nycprops_transformed.dtypes
```

```python
#identify and save the - character in the columns intended to be for integers
character=nycprops_transformed.iloc[1, 8]
character
```

```python
#eliminate - character in any cells so integer columns only have integers in them
nycprops_transformed=nycprops_transformed.replace(character, np.nan)
nycprops_transformed.head()
```
